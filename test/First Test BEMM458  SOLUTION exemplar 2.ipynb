## Question 2. Which city has the cleanest air ?
### Tip: consider adapting the following example: https://seaborn.pydata.org/examples/grouped_boxplot.html

```python
Cleanestcity = airQualityCities #Cleanestcity has been defined
```

```python
Cleanestcity # Print Cleanestcity
```

## Question 3. Which location is the most poluted in London ?

```python
citygrp1= airQualityCities.groupby('city')
```

```python
londoncities = citygrp1.get_group('London')
```

```python
londoncities
```

```python
sns.set(rc={'figure.figsize':(25,50)})
boxplot = sns.boxplot(x='location',y='value',hue = 'parameter',linewidth=None,data = londoncities)
boxplot.set(ylim=(0, 150))
```

```python
#Response London Harlington is the most poluted location in London
```

## Question 4. Plot the evolution of air polution accross time in locations in London
### Tip: Consider using Relplot: https://seaborn.pydata.org/generated/seaborn.relplot.html#seaborn.relplot

```python
sns.relplot( data=londoncities, x="date.local", y="value", col="location", hue="parameter", kind="line",
)
```

```python
#Response Evolution of air polution accross time in locations in London is according to the plot above.
```

## Question 5: Is there any relationship between levels of PM25 and PM10 in London?

```python
pmfilter = (londoncities['parameter'] == 'pm25')|(londoncities['parameter'] == 'pm10')
```

```python
pmdf = londoncities[pmfilter]
```

```python
sns.lineplot(data=pmdf , x='date.local', y='value', hue = 'parameter')
```

```python
#Response PM25 and PM10 in London are relatively correlated according to the plot above.
```
